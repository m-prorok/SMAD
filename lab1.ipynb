```python
import pandas as pd
covid_data = pd.read_csv("datasets\covid-data.csv")
```

```python
poland_covid_data=covid_data[covid_data['location']=='Poland']
```

```python
poland_covid_data['date']= pd.to_datetime(poland_covid_data['date'], format='%Y-%m-%d')
```

```python
poland_covid_data_weekly = poland_covid_data.groupby(['iso_code', 'continent', 'location', pd.Grouper(key='date', freq='W-MON')]).sum().reset_index()
```

```python
poland_covid_data_weekly.columns
```

```python
poland_covid_data_w_types = poland_covid_data_weekly.dtypes
nofloat = poland_covid_data_w_types != 'float64'
poland_covid_data_w_types[nofloat]
```

```python
value_columns=set(poland_covid_data_weekly.columns)-{'iso_code', 'continent', 'location', 'date', 'tests_units'}
```

```python
import matplotlib.pyplot as plt

for column in value_columns:
    poland_covid_data_weekly.plot('date',y=[column], title=column+'/date')
    plt.savefig(column+'_date.png')
```

```python
import copy

pol_cov = copy.deepcopy(poland_covid_data_weekly)
```

```python
pol_cov.dropna(axis=1, how='all', inplace=True)
pol_cov.info()
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

value_columns=list(set(pol_cov.columns)-{'iso_code', 'continent', 'location', 'date', 'tests_units'})
scaler = MinMaxScaler()
pol_cov[value_columns] = scaler.fit_transform(pol_cov[value_columns])
pol_cov
```

```python
import matplotlib.pyplot as plt

plt.plot(pol_cov['date'], pol_cov['new_cases']/pol_cov['new_cases'].shift(-1))
```
