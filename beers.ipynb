```python
beers.drop(columns = ['Name', 'Brewery', 'Description'] , inplace=True)
beers.info()
```

```python
beers.set_index(keys=['Style', 'Beer Name (Full)'])
```

```python
import matplotlib.pyplot as plt

value_beers = set(beers.columns)-{'Style', 'Beer Name (Full)'}

for col in value_beers:
    plt.boxplot(beers[col])
    plt.title(col)
    plt.show()
```

```python
import matplotlib.pyplot as plt

value_beers = set(beers.columns)-{'Style', 'Beer Name (Full)'}

for col in value_beers:
    plt.hist(beers[col])
    plt.title(col)
    plt.show()
```
